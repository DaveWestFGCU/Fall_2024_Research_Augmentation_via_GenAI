```python
dataset_name = "enISEAR"
llm_name = "Llama3.1 8B instruct-q8"
start_record_with = "I felt ... "
```

```python
import pandas as pd

from config_files import dataset_config
dataset = dataset_config.dataset[dataset_name]

from config_files import LLM_config
llm = LLM_config.model[llm_name]

dataframe = pd.read_parquet(f"../synthetic_datasets/{dataset["id"]}/{llm["id"].replace(":", "_")}.parquet" )

dataframe
```

```python
df_wrong = dataframe[~dataframe["text"].str.match('I felt ... ')]

df_wrong
```

```python
dataframe["text"] = df_wrong["text"].apply(lambda x: x.lstrip())
dataframe["text"] = df_wrong["text"].apply(lambda x: start_record_with + x)

dataframe
```

```python
dataframe.to_parquet(f"../synthetic_datasets/"
                     f"{dataset["id"]}/"
                     f"{llm["id"].replace(":", "_")}.parquet" )
```
